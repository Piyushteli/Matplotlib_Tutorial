# Stack Plot.

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[10,20,30,40,50]

plt.stackplot(x,y)
plt.show()
```

# Multiple stack Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[2,3,12,4,7]
area2=[2,5,4,6,3]
area3=[2,3,4,6,3]

plt.stackplot(x,area1,area2,area3)
plt.show()
```

# Parameter.

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[2,3,12,4,7]
area2=[2,5,4,6,3]
area3=[2,3,4,6,3]
l=["area1","area2","area3"]

plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","m"])

plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=2)
plt.show()
```
