# Scatter Plot.

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no,color="r")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","b","m",'g',"r","b","g"]
size=[100,200,300,400,500,600,700]
plt.scatter(day,no,color=colors,s=size,marker="*",edgecolor="g",linewidth=2)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","b","m",'g',"r","b","g"]
size=[100,200,300,400,500,600,700]
plt.scatter(day,no,color=colors,s=size)
t=plt.colorbar()
t.set_label("Colorbar")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```
