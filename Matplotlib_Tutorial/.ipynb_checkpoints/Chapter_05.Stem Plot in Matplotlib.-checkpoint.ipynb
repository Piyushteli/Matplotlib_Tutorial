# Stem Plot.

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[1,2,3,4,5]

plt.stem(x,y)
plt.show()
```

# Parameters.

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[3,4,5,6,5,6,7]

plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt='g',label="Python",use_line_collection=False)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[3,4,5,6,5,6,7]

plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt='g',label="Python",use_line_collection=False,orientation="horizontal")

plt.legend()
plt.show()
```
