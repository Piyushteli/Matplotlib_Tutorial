# Axis Matplotlib Plots.

How to Change Axis in Matplotlib.

# Functions.

1)xticks()

2)yticks()

3)xlim()

4)ylim()

5)axis()

```python
import matplotlib.pyplot as plt 
x=[1,2,3,4,5]
y=[3,4,2,5,2]

plt.plot(x,y)

plt.xticks(x)
plt.yticks(x)

plt.xlim(0,10)
plt.ylim(0,10)

plt.axis([0,10,0,7])

plt.show()
```
