# Text in Matplotlib Plots.

1)text

2)annotate

3)xlabel

4)ylabel

5)title

# 1)text

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.title("Course",fontsize=15)

plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.text(2,5,"Java",fontsize=15,style="italic",bbox={"facecolor":"r"})


plt.show()
```

# 2)Annotate

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.title("Course",fontsize=15)

plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.text(2,5,"Java",fontsize=15,style="italic",bbox={"facecolor":"r"})
plt.annotate("Python",xy=(2,1),xytext=(3,3),arrowprops=dict(facecolor="black",shrink=100))


plt.show()
```

# 3)legend()

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.title("Course",fontsize=15)

plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)


plt.legend(["UP"],loc=9,facecolor="r",edgecolor="yellow",framealpha=0.5,shadow=True)
plt.show()
```
