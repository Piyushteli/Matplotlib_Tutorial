# Step Plot in Matplotlib.


```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.step(x,y)
plt.show()
```

# Parameters.

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.step(x,y,color="r",marker="o",ms=10,mfc="g",label="Python")

plt.title("Python")
plt.xlabel("x-Axis")
plt.ylabel("y-Axis")

plt.legend()
plt.grid()
plt.show()
```
