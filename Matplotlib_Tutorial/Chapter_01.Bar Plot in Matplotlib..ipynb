# Bar Plot

```python
import matplotlib.pyplot as plt
x=["Python","C","C++","Java"]
y=[85,70,60,82]
z=[20,30,40,50]

plt.xlabel("Languages",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Course")
c=["y","b","m","g"]

plt.bar(x,y,width=0.4,color="r",label="Popularity")
plt.bar(x,z,width=0.4,color="y",label="Popularity1")

plt.legend()
plt.show()
```

# Vertical Bar Graph.

```python
import matplotlib.pyplot as plt
import numpy as np

x=["Python","C","C++","Java"]
y=[85,70,60,82]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Languages",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Course")


plt.bar(p,y,width,color="r",label="Popularity")
plt.bar(p1,z,width,color="y",label="Popularity1")

plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

# Horizontal Bar Graph.

```python
import matplotlib.pyplot as plt
x=["Python","C","C++","Java"]
y=[85,70,60,82]
z=[20,30,40,50]

plt.xlabel("Languages",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Course")


plt.barh(x,y,color="r",label="Popularity")
plt.barh(x,z,color="y",label="Popularity1")

plt.legend()
plt.show()
```
